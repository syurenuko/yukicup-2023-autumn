```python
import numpy as np
import pandas as pd
```

```python
train = pd.read_csv('yukiCup2023Autumn-atmaCup_dataset/yukiCup 2023 Autumn x atmaCup/train.csv')
test = pd.read_csv('yukiCup2023Autumn-atmaCup_dataset/yukiCup 2023 Autumn x atmaCup/test.csv')
era = pd.read_csv('yukiCup2023Autumn-atmaCup_dataset/yukiCup 2023 Autumn x atmaCup/era.csv')
sample_submission = pd.read_csv('yukiCup2023Autumn-atmaCup_dataset/yukiCup 2023 Autumn x atmaCup/sample_submission.csv')
```

```python
# 両データを「時代」カラムをキーにして結合
df_train = pd.merge(train, era, on='時代', how='left')
df_test = pd.merge(test, era, on='時代', how='left')
```

```python
df_train
```

```python
# 緯度と経度の掛け算
df_train['緯度*経度'] = df_train['緯度'] * df_train['経度']
df_test['緯度*経度'] = df_test['緯度'] * df_test['経度']

# 終了と開始の差分
df_train['終了-開始'] = df_train['終了'] - df_train['開始']
df_test['終了-開始'] = df_test['終了'] - df_test['開始']
```

```python
#categorical変数抽出
cat_cols = [c for c in df_test.columns if df_train[c].dtype == np.object]

# numerical変数抽出
num_cols = [c for c in df_test.columns if df_train[c].dtype != np.object]

#目的変数抽出
target_col = 'is_kokuhou'

print(cat_cols)
print(num_cols)
print(target_col)
```

```python
# LabelEncoding

for col in cat_cols:
    encoder = {c: i for i, c in enumerate(train[col].unique())}
    df_train[f'label_{col}'] = df_train[col].map(encoder)
    df_test[f'label_{col}'] = df_test[col].map(encoder)
```

```python
#　CountEncoding

for col in cat_cols:
    encoder = df_train[col].value_counts()
    df_train[f'count_{col}'] = df_train[col].map(encoder)
    df_test[f'count_{col}'] = df_test[col].map(encoder)
```

```python
# 集約特徴量
agg_cols = ['min', 'max', 'mean', 'std']

for col in cat_cols:
    grp_df = df_train.groupby(col)[num_cols].agg(agg_cols)
    grp_df.columns = [f'{col}_' + '_'.join(c) for c in grp_df.columns]
    df_train = df_train.merge(grp_df, on=col, how='left')
    df_test = df_test.merge(grp_df, on=col, how='left')
```

```python
df_train.to_csv('train_feature_ver1.csv', index=False)
df_test.to_csv('test_feature_ver1.csv', index=False)
```
