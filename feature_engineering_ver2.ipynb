```python
import numpy as np
import pandas as pd
```

```python
train = pd.read_csv('yukiCup2023Autumn-atmaCup_dataset/yukiCup 2023 Autumn x atmaCup/train.csv')
test = pd.read_csv('yukiCup2023Autumn-atmaCup_dataset/yukiCup 2023 Autumn x atmaCup/test.csv')
era = pd.read_csv('yukiCup2023Autumn-atmaCup_dataset/yukiCup 2023 Autumn x atmaCup/era.csv')
sample_submission = pd.read_csv('yukiCup2023Autumn-atmaCup_dataset/yukiCup 2023 Autumn x atmaCup/sample_submission.csv')
```

```python
# 両データを「時代」カラムをキーにして結合
df_train = pd.merge(train, era, on='時代', how='left')
df_test = pd.merge(test, era, on='時代', how='left')
```

```python
df_train
```

```python
# 緯度と経度の掛け算
df_train['緯度*経度'] = df_train['緯度'] * df_train['経度']
df_test['緯度*経度'] = df_test['緯度'] * df_test['経度']

# 終了と開始の差分
# df_train['終了-開始'] = df_train['終了'] - df_train['開始']
# df_test['終了-開始'] = df_test['終了'] - df_test['開始']
```

```python
#categorical変数抽出
cat_cols = [c for c in df_test.columns if df_train[c].dtype == np.object]

# numerical変数抽出
num_cols = [c for c in df_test.columns if df_train[c].dtype != np.object]

#目的変数抽出
target_col = 'is_kokuhou'

print(cat_cols)
print(num_cols)
print(target_col)
```

```python
# 結合するカラムのリスト
columns_to_combine = ['棟名', '種別2', '国', '時代', '都道府県 ※美工品は「所有者住所（所在都道府県）」', '所在地', '保管施設の名称', '所有者名', '管理団体又は責任者']

# カラムの結合
df_train['string'] = df_train.apply(lambda row: ' '.join([str(row[col]) for col in columns_to_combine]), axis=1)
df_test['string'] = df_test.apply(lambda row: ' '.join([str(row[col]) for col in columns_to_combine]), axis=1)

# 結果を確認
print(df_train[['string']].head())
```

```python
import tensorflow as tf
import tensorflow_text
import tensorflow_hub as hub

from tqdm import tqdm


tqdm.pandas()

embedder = hub.load(
    "https://tfhub.dev/google/universal-sentence-encoder-multilingual/3")
```

```python
df_train["string"]
```

```python
df_test["string"]
```

```python
features_train = np.stack(
    df_train["string"].fillna("").progress_apply(lambda x: embedder(x).numpy().reshape(-1)).values
)

features_test = np.stack(
    df_test["string"].fillna("").progress_apply(lambda x: embedder(x).numpy().reshape(-1)).values
)
```

```python
df_train = pd.concat([df_train, pd.DataFrame(features_train)], axis=1)
df_test = pd.concat([df_test, pd.DataFrame(features_test)], axis=1)
```

```python
df_train.to_csv('train_feature_ver2.csv', index=False)
df_test.to_csv('test_feature_ver2.csv', index=False)
```
